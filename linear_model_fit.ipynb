```python
# import libraries
import numpy as np # array
import matplotlib.pyplot as plt  # plotting
import astropy  # astropy
from astropy.modeling import models, fitting # model fitting
import random # for random number module
```

```python
# give x and y data
x = np.linspace(1, 100, 10)
y =  x + np.random.randint(1, 50, 10)  # add some noise
x, y
```

```python
# plot scatter
plt.scatter(x, y)
plt.show()
```

```python
# to intialize linear fitter
fit = fitting.LinearLSQFitter()
# initialize linear model
fit_line = models.Linear1D()
```

```python
# fitting the scatter plot

fitted_line = fit(fit_line, x, y)
plt.plot(x, fitted_line(x), c='red', lw= 3)
plt.scatter(x, y)
plt.show()
```
